```python
# Vamos a seguir trabajando con el csv de rating-and-performance, pero vamos a añadir un csv más para completar la información 
# de nuestro e-commerce. En concreto, vamos a incluir insight.csv.
# 1. Unid los dos DataFrames en uno solo, usando el método que considereis más correcto. Querremos quedarnos con todos los 
# productos que estén en los dos conjuntos de datos.

df = pd.read_csv("rating-and-performance.csv.csv", index_col=0)
df_insights = pd.read_csv("insight.csv.csv", index_col=0)

df.head(2)
```

```python
# 3. Algunas de las columnas tienen . en sus nombres y en otras tenemos _. Unifica el nombre de las columnas para que todas 
# tengan "_". Además, todas deben estar en minúsculas.

col_nuevas = {col : col.lower().replace(".", "_") for col in df_insights.columns}

df_merge.rename(columns= col_nuevas, inplace = True)

```

```python
df_merge
```

```python
# 4. En la columna de merchant_info_subtitle, si nos fijamos hay algunas celdas donde tenemos entre paréntesis el número total 
# de calificaciones para el comerciante (que es la misma información que la que tenemos en la columna "merchant_rating_count"). 
# Eliminad de la columna "merchant_info_subtitle" la parte del string de los paréntesis. Guardar el resultado de vuestra 
# limpieza en una columna que se llame igual, es decir, "merchant_info_subtitle".

df["merchant_info_subtitle"] = df_merge["merchant_info_subtitle"].str.replace(r'\(.*?\)', '', regex=True)
```

```python
df["merchant_info_subtitle"]
```

```python
# 5. Si nos fijamos, en la primera fila antes solo teníamos un paréntesis, pero los hemos eliminado en el paso anterior, y ahora
#  tenemos un string vacío, lo cual no tiene mucho sentido. En este ejercicio debereis reemplazar esos strings vacios por nulos 
# de Pandas (NOTA estos deben ser indicados de la siguiente forma np.nan, y deberéis importar numpy).

df["merchant_info_subtitle"] = df["merchant_info_subtitle"].replace("", np.nan, regex = True)
df.head(3)
```

```python
# 6. Hay ciertas columnas que no vamos a necesitar para nuestro análisis. En concreto, vamos a eliminar las columnas de 
# badges_count, product_variation_inventory, urgency_text, merchant_title, merchant_has_profile_picture, 
# merchant_profile_picture, product_picture, total_urgency_count, has_urgency_banner y urgency_text_rate. Guardad los cambios 
# realizados hasta este punto en un csv para usarlo en los ejercicios de la lección de mañana.

eliminar = ["badges_count", "product_variation_inventory",'has_urgency_banner', 'merchant_title', 'merchant_has_profile_picture',  'merchant_profile_picture', 'product_picture']
df.drop(eliminar, axis = 1, inplace = True)
```
